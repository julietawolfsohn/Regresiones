```python
def grafico_out_boxplot(df):
    cols = df.select_dtypes(include=['number']).columns

    filas = len(cols)
    columnas = 1

    fig, axes = plt.subplots(filas, columnas, figsize=(6,50 ))

    for i, columna in enumerate(cols):
        sns.boxplot(y=df[columna], ax=axes[i])
        axes[i].set_title(columna)

    plt.tight_layout()
    return (plt.show())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_histograms(dataframe):
    # Obtén la lista de nombres de todas las columnas del DataFrame
    column_names = dataframe.columns
    

    # Itera a través de los nombres de las columnas y crea un histograma para cada una
    for column_name in column_names:
        plt.figure(figsize=(4, 2))  # Tamaño de la figura
        sns.histplot(dataframe[column_name], kde=True, edgecolor='w')
        plt.title(f'Histograma de {column_name}')
        plt.xlabel(column_name)
        plt.ylabel('Frecuencia')
        plt.show()

# Llama a la función con tu DataFrame (emprendimientos en este caso)
plot_histograms(dff)
```


El error promedio y el error elevado al cuadrado son utilizan para evaluar la calidad del rendimiento de un modelo de regresión o de predicción. Miden las predicciones de un modelo y los valores reales de un conjunto de datos. 

Error Promedio --> mide el promedio de las diferencias absolutas entre las predicciones de un modelo y los valores reales. Cuanto menor sea el valor del MAE, mejor será el rendimiento del modelo, ya que indicaría que las predicciones están más cerca de los valores reales en promedio.
Se calcula haciendo -->  la diferencia absoluta entre cada predicción y su valor real. Luego, se promedian estas diferencias absolutas para obtener el MAE.

Error Elevado al Cuadrado --> mide el promedio de las diferencias al cuadrado entre las predicciones de un modelo y los valores reales. El MSE es más sensible a errores grandes, ya que penaliza más fuertemente las discrepancias grandes en comparación con el MAE.
Se calcula haciendo --> la diferencia al cuadrado entre cada predicción y su correspondiente valor real. Luego, se promedian estas diferencias al cuadrado para obtener el MSE.

E. Analice del Sector de la Industria en el Beneficio Neto.


```python

beneficio_promedio_por_sector = dff.groupby('Industry_Sector')['Net_Profit'].mean()
beneficio_total_por_sector = dff.groupby('Industry_Sector')['Net_Profit'].sum()
max_beneficio_sector = beneficio_total_por_sector.idxmax() # idxmax --> se utiliza para encontrar el índice de la fila que contiene el valor máximo en una Serie o DataFrame

print("Beneficio neto promedio por sector:")
print(beneficio_promedio_por_sector)

print("\nBeneficio neto total por sector:")
print(beneficio_total_por_sector)

print(f"\nEl sector con el beneficio neto máximo es: {max_beneficio_sector}")
```

```python
import matplotlib.pyplot as plt

# Datos del beneficio neto por sector
data = [dff[dff['Industry_Sector'] == sector]['Net_Profit'] for sector in dff['Industry_Sector'].unique()]

# Configurar el tamaño del gráfico
plt.figure(figsize=(10, 6))
sns.set_palette("Set1")
boxplot = plt.boxplot(data, labels=dff['Industry_Sector'].unique(), patch_artist=True)

# Configuración adicional
plt.xlabel('Sector')
plt.ylabel('Beneficio Neto')
plt.title('Boxplot del Beneficio Neto por Sector')

# Mostrar el gráfico
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

"Tech" tiene el beneficio neto máximo ya que entre todos los sectores es el que tiene la suma mayor de  beneficio neto. Es el que genero más ingresos netos en comparación con otros sectores.

Su Beneficio neto promedio es de 119788.948734, su beneficio neto total es de 2.156201e+06. Por lo tanto, el sector con el beneficio neto máximo es: Tech.
